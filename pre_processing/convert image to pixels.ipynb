```python
from pathlib import Path
import numpy as np
import pandas as pd
import os
```

## Load the data set

```python
data = pd.read_csv('../../DATA/TEST/Images/test_labels.csv')
data.head()
```

## Download the images & preparation

```python
from skimage import io

images = []
labels = []
#Label : 0 for bees & 1 for wasps

for index, row in data.iterrows():
    image_path = '../../DATA/TEST/Images/' + row['label'] + '/' + row['image_name']
    label = row['label']
    images.append(io.imread(image_path))
    labels.append(label)
```

Now we check if all images are squared images :

```python
# for i in range(0,len(images)):
#     if(np.shape(images[i])[0] != np.shape(images[i])[1]):
#         print(np.shape(images[0])[0],np.shape(images)[1])
#         print("Erreur, des images ont mal été modifiés :" , i)
```

Now we will flatten the array to insert it into the dataframe :

```python
flatten_images = []
for image in images:
    flatten_images.append(image.ravel())
```

Now we will convert the array img into a dataframe :

```python
total_columns = len(flatten_images[0]) # 128 x 128 x 3
resolution = 128
```

```python
columns = []
for i in range(0,resolution*resolution):
    name = 'pixel_'+str(1+int(i/resolution))+'_'+str(1+(i%resolution))
    columns.append(name+'_R')
    columns.append(name+'_G')
    columns.append(name+'_B')
```

```python
columns[-1]
```

```python
labels_dataframe = pd.DataFrame(columns=['label'], data=labels)
to_be_merged_dataframe = pd.DataFrame(columns = columns, data=flatten_images)    
final_dataframe = to_be_merged_dataframe.merge(labels_dataframe, left_index=True, right_index=True)
```
